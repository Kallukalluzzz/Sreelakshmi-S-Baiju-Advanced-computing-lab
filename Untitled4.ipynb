```python
import pandas as pd
data=pd.read_csv('/content/airline-passengers.csv',index_col='Month')
data
```

**PLOT THE DATA**

```python
from matplotlib import pyplot
data.plot(figsize=(14,8))
pyplot.title("Daily temperature")
pyplot.show()
```

**PLOT HISTOGRAM**

```python
data.hist()
pyplot.show()
```

**PLOT ACF**

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(data)
pyplot.show()
```

**PLOT PACF**

```python
from statsmodels.graphics.tsaplots import plot_pacf
plot_pacf(data)
pyplot.show()
```

**STEPS FOR CHECKING STATIONARITY**

```python
x = data.values
split = round(len(x) / 2)
x1, x2 = x[0:split], x[split:]
mean1, mean2 = x1.mean(), x2.mean()
var1, var2 = x1.var(), x2.var()
```

```python
print(mean1, mean2)
```

```python
print(var1, var2)
```

**LOG TRANSFORM**

```python
from numpy import log
x = data.values
x = log(x)
pyplot.hist(x)
pyplot.show()
pyplot.plot(x)
pyplot.show()
```

```python
x = data.values
x = log(x)
split = round(len(x) / 2)
x1, x2 = x[0:split], x[split:]
mean1, mean2 = x1.mean(), x2.mean()
var1, var2 = x1.var(), x2.var()
```

```python
print(mean1, mean2)
```

```python
print(var1, var2)
```
